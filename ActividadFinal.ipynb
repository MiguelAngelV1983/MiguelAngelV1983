```python
import pandas as pd
import numpy as np

#Importamos los valores de los csv y los metemos en un dataset conjunto indicando el tipo de vino que son (0: tinto, 1: blanco)
red_wines = pd.read_csv('winequality-red.csv', sep=';')
red_wines['type'] = np.zeros_like(np.arange(len(red_wines.index)))
#print(red_wines)


white_wines = pd.read_csv('winequality-white.csv', sep=';')
white_wines['type'] = np.ones_like(np.arange(len(white_wines.index)))
#print(white_wines)


wines = pd.DataFrame([], columns=red_wines.columns)
for i in range(len(red_wines.index)):
    wines.loc[i] = red_wines.loc[i]

for j in range(1, len(white_wines.index)+1):
    wines.loc[i+j] = white_wines.loc[j-1]

print(wines)
```

```python

import matplotlib.pyplot as plt

#imprimimos un gr√°fico que represente el total de vinos por calidad

qualities = list(map(int,wines['quality'].sort_values().unique()))
wines_per_quality = []
for i in range(len(qualities)):
  wines_per_quality.append(len(wines[wines['quality'] == qualities[i]]))

print(qualities)
print(wines_per_quality)

fig, ax = plt.subplots()
fig.set_figheight(6)
fig.set_figwidth(20)
rects1 = ax.bar(np.arange(len(qualities)), wines_per_quality, width=0.75, label=None)
#rects1..set_color('#ff9757')
for nbar in range(len(qualities)):
    rects1[nbar].set_color(('#ff9757'))

ax.set_ylabel('Numero de Vinos')
ax.set_title('Cantidad de Vinos por Calidad')
ax.set_xticks(np.arange(len(qualities)))
ax.set_xticklabels(qualities)
ax.set_xlabel('Calidad del Vino')


fig.tight_layout()

rects = ax.patches

labels = [f"{wines_per_quality[i]}" for i in range(len(rects))]

for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height + 5, label, ha="center", va="bottom")

plt.plot()
```

```python
from scipy import stats

#Media
mean_quality = np.mean(wines['quality'])
#print(mean_quality)
#Mediana
median_quality = np.median(wines['quality'])
#print(median_quality)
#Moda
mode_quality = stats.mode(wines['quality'])
#print(mode_quality)
#Varianza
var_quality = np.var(wines['quality'])
#print(var_quality)
#Desviacion Tipica
std_quality = np.std(wines['quality'])
#print(std_quality)

print(wines['quality'].max())
display(wines[wines['quality'] == wines['quality'].max()])

print(wines['quality'].min())
display(wines[wines['quality'] == wines['quality'].min()])

```

```python
#comparativa vinos blancos
    #mayor y menor calidad
    #total de cada tipo (calidad < y >)
    #Gr√°fica comparativa de ambas calidades
    
wh_wines_max_qua = wines[wines['quality'] == wines['quality'].max()]
wh_wines_min_qua = wines[wines['quality'] == wines['quality'].min()]
wh_wines_min_qua = wh_wines_min_qua[wh_wines_min_qua['type'] == 1]

display(wh_wines_max_qua.columns)
#display(wh_wines_min_qua)

#atributes = ['Media', 'Mediana', 'Moda', 'Des. Tip', 'Varianza']
#atr_wh_maxq = [np.mean(wh_wines_max_qua), np.median(wh_wines_max_qua), np.mode(wh_wines_max_qua)[0][0]]
print(np.mean(wh_wines_max_qua)[0])
```

```python
x = np.arange(len(wh_wines_max_qua.loc[:,['free sulfur dioxide', 'total sulfur dioxide']].columns))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
fig.set_figheight(6)
fig.set_figwidth(20)
rects1 = ax.barh(x - width/2, np.mean(wh_wines_max_qua.loc[:,['free sulfur dioxide', 'total sulfur dioxide']]), width, label='Mayor Calidad')
rects2 = ax.barh(x + width/2, np.mean(wh_wines_min_qua.loc[:,['free sulfur dioxide', 'total sulfur dioxide']]), width, label='Menor Calidad')

# Add some text for labels, title and custom x-axis tick labels, etc.
#ax.set_ylabel('Deportistas')
ax.set_title('Media de los atributos de los vinos')
ax.set_yticks(x)
ax.set_yticklabels(wh_wines_max_qua.loc[:,['free sulfur dioxide', 'total sulfur dioxide']].columns)
ax.set_ylabel('Atributos de los vinos')
ax.legend()

for index, value in enumerate(np.mean(wh_wines_max_qua.loc[:,['free sulfur dioxide', 'total sulfur dioxide']])):

    plt.text(value, index-width/2, str(value))

for index, value in enumerate(np.mean(wh_wines_min_qua.loc[:,['free sulfur dioxide', 'total sulfur dioxide']])):

    plt.text(value, index+width/2, str(value))

fig.tight_layout()

plt.show()
```
